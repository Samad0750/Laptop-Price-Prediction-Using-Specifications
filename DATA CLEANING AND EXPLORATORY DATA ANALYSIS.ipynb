```python
import re
```

```python
def extract_frequency(cpu_string):
    match = re.search(r'\d+(\.\d+)?GHz', cpu_string)
    return match.group(0) if match else None

# Apply the function to create the new column
df['Frequency'] = df['CPU'].apply(extract_frequency)

```
